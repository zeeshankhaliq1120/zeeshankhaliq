```python
## Loading packages:

import numpy as np
import pandas as pd 
import csv


# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
## Data Extraction:

try:   #Try/except function is used to check for errors of not finding the file when extracting and alertz the user when such error occurs
    with open('/kaggle/input/data-science-and-stem-salaries/Levels_Fyi_Salary_Data.csv') as Ds_STEM_Data:
        CsvReader = csv.reader(Ds_STEM_Data) #reading as CSV file
        for row in CsvReader: # for every row in incoming extracted data, print the row
            print(row)
except OSError as err: #print error message when file not found error occurs
    print("Error: File not found")

## Data Transformation:
try: #Try/except function is used again to account for possible file not found errors when converting to dataframe
    Ds_STEM = pd.read_csv('/kaggle/input/data-science-and-stem-salaries/Levels_Fyi_Salary_Data.csv')
except OSError as err: #print error message when file not found error occurs
    print('Error: File not found')

Ds_STEM_salaries = Ds_STEM[['company', 'title', 'location', 'basesalary']] #limiting to columns of interest

nyc_salaries = Ds_STEM_salaries[Ds_STEM_salaries.location == 'New York, NY'] #Filtering to only include salaries in New York City

PM_nyc_salaries = nyc_salaries[nyc_salaries.title == 'Product Manager'] #Filtering to only include salaries of Product Manager title in NYC

PM_nyc_salaries #outputing data frame, contains 336 rows and 4 columns

## Loading to JSON File:

import json #converting from CSV to JSON
final = PM_nyc_salaries.to_json(orient='split') #making use of dataframe.to_json function
final_parsed = json.loads(final) #parsing

json.dumps(final_parsed) #returning JSON string representation
```
